```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) plot the function 1/x between 1 and 5 using 100 support points!
x = np.linspace(1,5,100)
y = 1/x
plt.plot(x,y)
plt.show()
```

```python
# b) generate a scatter plot of the same data as in a)!
plt.scatter(x,y)
plt.show()
```

```python
# c) generate a bar plot of the same data as in a)!
plt.bar(x,y)
plt.show()
```

```python
# d) plot 1/x and âˆšx together in a single scatter plot, same range as before
y1 = np.sqrt(x)
plt.scatter(x,y)
plt.scatter(x,y1)
plt.show()
```
