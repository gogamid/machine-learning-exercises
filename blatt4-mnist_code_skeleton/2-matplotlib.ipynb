```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) plot the function 1/x between 1 and 5 using 100 support points!
x = np.linspace(1,5,100)
y = 1/x
plt.plot(x,y)
plt.show()
```

```python
# b) generate a scatter plot of the same data as in a)!
plt.scatter(x,y)
plt.show()
```

```python
# c) generate a bar plot of the same data as in a)!
plt.bar(x,y)
plt.show()
```

```python
# d) plot 1/x and √x together in a single scatter plot, same range as before
y1 = np.sqrt(x)
plt.scatter(x,y)
plt.scatter(x,y1)
plt.show()
```

```python
#a) plot the function x3 between 0 and 2 using precisely 100 support points!
x = np.linspace(0,2,100)
y = x**3
plt.plot(x,y)
plt.show()
```

```python
#b) generate 9 random "images" of shape (20,20) with values between 0 and 1, and display them together in a single window!
images = np.random.rand(9,20,20)
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i])
plt.show()
```
