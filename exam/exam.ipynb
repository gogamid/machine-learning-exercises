```python
#2.1 basics and slicing
import numpy as np
#Create an array X with random entries between 0 and 1, of shape (10,5,5,3). We will assume that this represents a stack of RGB image samples.
x = np.random.random((10,5,5,3))
print(x)

# Set the 3 lowermost rows of the first sample to -1
x[0, -3:] = -1
print(x)

#Slice out the 4th sample to a new array W, but keeping only every 2nd row!
w = x[3, ::2]

#copy out samples with indices 0,1,3 to a new array Z using a single assignment operation!
z = x[[0,1,3]]

#copy out all samples whose sum is smaller than 5 into a new arrav M
m = x[x.sum(axis=(1,2,3)) < 5]
```

```python
#3.1
import tensorflow as tf
def S(X):
    return tf.reduce_sum(tf.square(X), axis=1)


X = tf.constant(np.array([[-1, -1, 5], [1, 1, 3], [0, 0, 1]]), dtype=tf.float32)
print(S(X))
```

```python
#3.2 a
# Implement f(x) = sumi(xi) as a Python function expecting a tensor of type tf. Variable as its only parameter and compute its output for the inputs at = (1, 2, 3) and a2 = (2, 0, 2).
def f(x):
    return tf.reduce_sum(x)

a1 = tf.Variable([1, 2, 3], dtype=tf.float32)
a2 = tf.Variable([2, 0, 2], dtype=tf.float32)
print(f(a1))
print(f(a2))

#3.2 b
# Use TF to compute iand display the value of Vf, evaluated for ~ = &, and for I = dr
with tf.GradientTape() as tape:
    x = tf.Variable(a1)
    y = f(x)
print(tape.gradient(y, x))

with tf.GradientTape() as tape:
    x1 = tf.Variable(a2)
    y1 = f(x1)
print(tape.gradient(y1, x1))
```
